```python
from onnx_tf.backend import prepare
import onnx

onnx_model_path = '오닉스 파일경로'
tf_model_path = '저장할 텐서플로우 폴더 경로'

onnx_model = onnx.load(onnx_model_path)
tf_rep = prepare(onnx_model)
tf_rep.export_graph(tf_model_path)
```

```python
import tensorflow as tf

saved_model_dir = '저장한 텐서플로우 폴더 경로 '
tflite_model_path = '저장할 tflite 파일 경로'

# Convert the model
converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_dir)
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS,tf.lite.OpsSet.SELECT_TF_OPS]
converter.optimizations = [tf.lite.Optimize.DEFAULT] # 동적범위 양자화 
#converter.target_spec.supported_types = [tf.float16] # float 16 양자화 
converter.experimental_enable_resource_variables = True    
tflite_model = converter.convert()

# Save the model
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)
```
